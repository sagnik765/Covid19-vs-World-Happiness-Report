```python
df = dataset.drop(["Lat","Long"], axis = 1, inplace = True)
```

Aggregate rows by country

```python
covid_dataset = dataset.groupby("Country/Region").sum()
```

Calculating First Derivative

```python
covid_dataset.loc["China"].diff().plot()
```

Maximum Infection Rate

```python
covid_dataset.loc["China"].diff().max()
```

```python
covid_dataset.loc["India"].diff().max()
```

```python
covid_dataset.loc["Japan"].diff().max()
```

```python
countries = list(covid_dataset.index)
max_infection_rates = []

for c in countries:
  max_infection_rates.append(covid_dataset.loc[c].diff().max())
covid_dataset["Max_Infection_Rates"] = max_infection_rates
```

```python
covid_dataset
```

Dataframe

```python
corona_data = pd.DataFrame(covid_dataset["Max_Infection_Rates"])
```

```python
corona_data
```
